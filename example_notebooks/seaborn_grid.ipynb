```python
import os 
import sys 
import time
import seaborn as sns
import patchworklib as pw 
print(pw.__version__ )
from functools import reduce
pw.overwrite_axisgrid() #When you use pw.load_seagorngrid, 'overwrite_axisgrid' should be executed.
print(pw.__version__, sns.__version__)
```

```python
df  = sns.load_dataset("penguins")
g1  = pw.load_seaborngrid(sns.pairplot(df, hue="species"))
g1.savefig()
```

```python
g1.move_legend("upper right", bbox_to_anchor=(0.28,1.01))
g1.savefig()
```

```python
planets = sns.load_dataset("planets")
cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
g2 = sns.relplot(
    data=planets,
    x="distance", y="orbital_period",
    hue="year", size="mass",
    palette=cmap, sizes=(10, 200),
)
g2.set(xscale="log", yscale="log")
g2.ax.xaxis.grid(True, "minor", linewidth=.25)
g2.ax.yaxis.grid(True, "minor", linewidth=.25)
g2.despine(left=True, bottom=True)
g2 = pw.load_seaborngrid(g2)
g2.savefig()
```

```python
# Load the penguins dataset
# Show the joint distribution using kernel density estimation
g3 = sns.jointplot(
    data=df,
    x="bill_length_mm", y="bill_depth_mm", hue="species",
    kind="kde",
)
g3 = pw.load_seaborngrid(g3, labels=["joint","marg_x","marg_y"])
g3.savefig()
```

```python
(g2/g3).savefig()
```
