```python
import sys
import seaborn as sns
import patchworklib as pw
```

```python
ax1 = pw.Brick(figsize=(2,2))
ax2 = pw.Brick(figsize=(1,2))
ax3 = pw.Brick(figsize=(2,1))
```

```python
ax1.savefig()
```

```python
ax2.savefig()
```

```python
ax3.savefig()
```

```python
ax13 = pw.hstack(ax1,ax3, adjust_width=False, adjust_height=False, va="b")
ax13.savefig()
```

```python
ax13 = pw.hstack(ax1,ax3, adjust_width=False, adjust_height=False, va="t")
ax13.savefig()
```

```python
ax12 = pw.vstack(ax1, ax2, adjust_width=False, adjust_height=False, ha="l", direction="b")
ax12.savefig()
```

```python
(ax12[ax1]|ax3).savefig()
```
