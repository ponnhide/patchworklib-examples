```python
import sys 
sys.path = ["/Users/hideto/dropbox/HIDETO_MORI.LAB/Experiments/Project/Patchplot/patchworklib"] + sys.path
import patchworklib as pw
```

```python
ax_list1 = []
for i in range(5):
    ax_list1.append(pw.Brick(figsize=(2,2), label="ax{}".format(i)))
ax_list2 = []
for i in range(5,10):
    ax_list2.append(pw.Brick(figsize=(2,2), label="ax{}".format(i)))

row1 = pw.stack(ax_list1, operator="|", margin=0.2)
row2 = pw.stack(ax_list2, operator="|", margin=0.2)
stacked_axes = row1/row2 
stacked_axes.savefig("../stack_example.png")
```

```python
ax10 = pw.Brick(figsize=(2,2), label="ax10")
stacked_axes = stacked_axes["ax9"]/ax10
stacked_axes.savefig()
```
