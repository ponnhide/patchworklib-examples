```python
import sys 
sys.path = ["/Users/hideto/dropbox/HIDETO_MORI.LAB/Experiments/Project/Patchplot/patchworklib"] + sys.path
import patchworklib as pw
```

```python
ax_list1 = []
for i in range(4):
    ax = pw.Brick(figsize=(2,2), label="ax{}".format(i))
    ax.set_xticks([0,50000,100000])
    ax.set_xticklabels(["0","50000","100000"], rotation=90)
    ax.set_ylim(0,100000)
    ax.set_xlabel("fizz")
    ax.set_ylabel("buzz")
    ax_list1.append(ax)

ax_list2 = []
for i in range(4,8):
    ax = pw.Brick(figsize=(2,2), label="ax{}".format(i))
    ax.set_xticks([0,50000,100000])
    ax.set_xticklabels(["0","50000","100000"], rotation=90)
    ax.set_ylim(0,100000)
    ax.set_xlabel("fizz")
    ax.set_ylabel("buzz")
    ax_list2.append(ax)

ax_list3 = []
for i in range(8,12):
    ax = pw.Brick(figsize=(2,2), label="ax{}".format(i))
    ax.set_xticks([0,50000,100000])
    ax.set_xticklabels(["0","50000","100000"], rotation=90)
    ax.set_ylim(0,100000)
    ax.set_xlabel("fizz")
    ax.set_ylabel("buzz")
    ax_list3.append(ax)

row1 = pw.stack(ax_list1, operator="|", margin=0.2)
row2 = pw.stack(ax_list2, operator="|", margin=0.2)
row3 = pw.stack(ax_list3, operator="|", margin=0.2)
stacked_axes = row1/row2/row3
stacked_axes.savefig()
```

```python
stacked_axes.set_supxlabel("sup_fizz")
stacked_axes.set_supylabel("sub_buzz")
stacked_axes.set_suptitle("fizzbuzz")
stacked_axes.set_index("A", fontsize=20)
stacked_axes._case.set_xlabel("hoge")
```

```python
stacked_axes.savefig()
```
